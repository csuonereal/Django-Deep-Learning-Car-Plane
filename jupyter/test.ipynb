```python
import numpy as np
from keras.models import load_model
from keras.preprocessing.image import load_img
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')

model = load_model('C:\\Users\\user\\Desktop\\DL car or plane')

plt.figure(figsize=(10,10))
for i in range(25):
    image = load_img(f'mix/{i+1}.jpg', target_size=(224, 224))
    img = np.array(image) 
    img = img.reshape(1,224,224,3)
    prediction = model.predict_classes(img)
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(image)
    
    name=""
    if prediction==1:
        name='plane'
    else:
        name='car'
    plt.xlabel(name)
plt.show()
```
